```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers
tickers = ['VIXM', 'TSLL', 'TSLS', 'SPY']

# Define the start and end date
start_date = '2022-08-01'
end_date = '2023-05-01'

# Fetch the data for each ticker
data = {}
for ticker in tickers:
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    data[ticker] = stock_data

# Save the data to CSV files
for ticker in tickers:
    data[ticker].to_csv(f"{ticker}_historical_data.csv")

# Visualize the closing prices for each ticker
plt.figure(figsize=(14, 7))
for ticker in tickers:
    plt.plot(data[ticker].index, data[ticker]['Close'], label=ticker)

plt.title('Closing Prices of Selected Tickers')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

print("Data downloaded, saved to CSV files, and visualized.")
```
